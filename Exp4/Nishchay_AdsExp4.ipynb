```python
# Define bins for classification
bins = [0, 10000, 20000, 30000, 40000, 50000, float('inf')]
labels = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5', 'Class 6']
data['charge_class'] = pd.cut(data['charges'], bins=bins, labels=labels)

# Split features and target
features = data.drop(['charges', 'charge_class'], axis='columns')
target = data["charge_class"]

# Perform one-hot encoding if necessary
new_features = pd.get_dummies(features, drop_first=True)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(new_features, target, test_size=0.25, random_state=0)

# Initialize and train BaggingClassifier model
model = BaggingClassifier()
model.fit(x_train, y_train)

# Predict on the test set
y_pred = model.predict(x_test)
```

```python
# Calculate classification metrics
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of this model is: ", round(accuracy * 100, 2), "%")

precision = precision_score(y_test, y_pred, average='weighted')
print("Precision of this model is= ", round(precision, 2))

recall = recall_score(y_test, y_pred, average='weighted')
print("Recall score of this model is= ", round(recall, 2))

f1 = f1_score(y_test, y_pred, average='weighted')
print("F1 Score of this model is= ", round(f1, 2))

roc_auc = roc_auc_score(y_test, model.predict_proba(x_test), multi_class='ovr', average='weighted')
print("ROC AUC of this model is= ", round(roc_auc, 2))

```
